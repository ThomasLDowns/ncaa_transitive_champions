```python
import re
```

```python
def build_they_beat_dict(filename):    
    txt = open(filename, 'r')

    for line in txt.readlines():
        line = line.replace('@', ' ').strip() # replace the at-home marker with a space
        line = re.sub(r'\s{2,}', '#', line)   # replace all white_spaces of 2 or more with a #
        cols = line.split("#")                # split on #
        teams = (cols[1], cols[3])
        scores = (int(cols[2]), int(cols[4]) if len(cols[4]) <= 3 else int(cols[4].split()[0]))
        i = 0 if scores[0] > scores[1] else 1
        j = (i + 1) % 2
        # add team to set of all teams
        for team in teams:
            all_teams.add(team)
        # add winning teams to dictionary with key of losing team
        if teams[j] not in they_beat.keys():
            they_beat[teams[j]] = set()
        they_beat[teams[j]].add(teams[i])
```

```python
def build_trans_champs_set(champ):    
    trans_champs = they_beat[champ].copy()
    left_to_check = they_beat[champ].copy()
    
    while len(left_to_check) >= 1:
        t = left_to_check.pop()
        if t in they_beat.keys():
            left_to_check |= (they_beat[t] - trans_champs)
            print(f"{t} also lost to: {', '.join(list(they_beat[t] - trans_champs))}")
            trans_champs |= they_beat[t]
            
    return trans_champs
```

```python
all_teams = set()
they_beat = dict()
build_they_beat_dict('ncaa_m_season.txt')
trans_champs = build_trans_champs_set('Virginia')
```

```python
print(', '.join(list(all_teams - trans_champs)))
```

```python
print(f"{len(all_teams)} total teams, {len(trans_champs)} of which were transitive champions")
```
